# Using Camelot

Let's import it and extract a table from a PDF.

```python
import camelot
```

```python
tables = camelot.read_pdf('players.pdf', flavor='stream', pages='1-end')
tables[0].df
```

## pdfplumber

If you're interested in Camelot, I want to say that [pdfplumber](https://github.com/jsvine/pdfplumber) is an absolute delight. It's a little more technical, but if that's your thing I'll link it here, too.

```python
import pdfplumber
import pandas as pd

table_settings = {/
    "vertical_strategy": "text",
    "horizontal_strategy": "text",
}

pdf = pdfplumber.open("players.pdf")

table = pdf.pages[0].extract_table(table_settings)
pd.DataFrame(table[1::], columns=table[0])
```
